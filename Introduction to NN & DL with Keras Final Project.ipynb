<h2> Data Preparation<h2>

```python
#Import libraries to prep data
import numpy as np
import pandas as pd
```

```python
#Load dataset
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.head()
```

```python
#Display dimension
concrete_data.shape
```

```python
#Check for missing values
concrete_data.isnull().sum()
```

```python
#Split data into features and target values
concrete_columns = concrete_data.columns

features = concrete_data[concrete_columns[concrete_columns != 'Strength']]
target = concrete_data['Strength']
```

```python
features.head()
```

```python
target.head()
```

<h2>A. Build a baseline model</h2>


```python
#Import Keras
import keras
from keras.models import Sequential
from keras.layers import Dense

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#Build Neural Network for Regression
def regression():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(features.shape[1],)))
    model.add(Dense(1))
    
    #Define optimizer and loss function
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    
    return model
    
```

```python
#Create model
reg_model = regression()
```

```python
mse_list = []

#50 iterations
for i in range(50):
    
    #Train Test Split
    feat_train, feat_test, target_train, target_test = train_test_split( features, target, test_size=0.3, random_state=4)
    
    #Train model
    reg_model.fit(feat_train, target_train, epochs=50)
    
    #Prediction on test data
    target_pred = reg_model.predict(feat_test)
    
    #Compute MSE for evaluation by comparing prediction and actual value
    mse = mean_squared_error(target_test, target_pred)
    mse_list.append(mse)
    print("Mean Squared Error:", mse)    
```

```python
#Calculate Mean and Satndard Deviation
mse_mean = np.mean(mse_list)
print("Mean of the MSE: ", mse_mean )

std = np.std(mse_list)
print("Standard Deviation of the MSE: ", std)
```

<h2>B. Normalize the data<h2>

```python
#Normalize data
feat_norm = (features - features.mean()) / features.std()
feat_norm.head()
```

```python
mse_list = []

#50 iterations
for i in range(50):
    
    #Train Test Split
    feat_train, feat_test, target_train, target_test = train_test_split( feat_norm, target, test_size=0.3, random_state=4)
    
    #Train model
    reg_model.fit(feat_train, target_train, epochs=50)
    
    #Prediction on test data
    target_pred = reg_model.predict(feat_test)
    
    #Compute MSE for evaluation by comparing prediction and actual value
    mse = mean_squared_error(target_test, target_pred)
    mse_list.append(mse)
    print("Mean Squared Error:", mse)    
```

```python
#Calculate Mean and Satndard Deviation
mse_mean = np.mean(mse_list)
print("Mean of the MSE: ", mse_mean )

std = np.std(mse_list)
print("Standard Deviation of the MSE: ", std)
```

Compared to A the mean of the MSEs decresed marginally (from 56.999975342284195 to 52.00467639929596), whereas there was a significant increse of the dtandard deviation (from 21.093072070768702 to 30.262179807987287).

<h1>D. Increase the number of hidden layers</h1>

```python
#Build Neural Network for Regression
def regression2():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(features.shape[1],)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    #Define optimizer and loss function
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    
    return model
    
```

```python
#Create model
reg_model2 = regression2()
```

```python
mse_list = []

#50 iterations
for i in range(50):
    
    #Train Test Split
    feat_train, feat_test, target_train, target_test = train_test_split( feat_norm, target, test_size=0.3, random_state=4)
    
    #Train model
    reg_model2.fit(feat_train, target_train, epochs=50)
    
    #Prediction on test data
    target_pred = reg_model2.predict(feat_test)
    
    #Compute MSE for evaluation by comparing prediction and actual value
    mse = mean_squared_error(target_test, target_pred)
    mse_list.append(mse)
    print("Mean Squared Error:", mse)    
```

```python
#Calculate Mean and Satndard Deviation
mse_mean = np.mean(mse_list)
print("Mean of the MSE: ", mse_mean )

std = np.std(mse_list)
print("Standard Deviation of the MSE: ", std)
```
